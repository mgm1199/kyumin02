# 2번 문제

```python
s = 'Hello World and Python'

str1 = ' '.join(s.split()[::-1])
print(str1)

str2 = ''.join(s.split()[::-1])
print(str2)
```

str1 : 우선 문자열 s 를 공백을 기준으로 나눈 리스트를, 역슬라이싱하여, 반대로 반환한다. 이것을 ' '을 사이에 넣어 join 함으로써, 단어가 뒤집한 문자열을 얻는다.

str2 : 방법은 동일하지만, join 메소드를 호출할때에, 사이에 공백이 아닌 아무것도 넣지 않음으로써 공백을 제거한 문자열을 얻는다.

# 3번 문제

b는 리스트 객체a 를 3번 나열하여 새로 생성된 리스트 객체이며, c는 리스트 객체 a를 또 리스트로 취하여 3번 나열한 리스트 객체이다. a를 참조하고 있기 때문에, 객체 a의 원소가 변경되면, c의 원소가 a이므로 c 또한 변경된다.

# 4번 문제

```python
s = '  first star   :   second star   :    third star  '
l = []

for i in s.split(':'):
    l.append(i.strip())
print(l)

l = [i.strip() for i in s.split(':')]
print(l)
```

1) for문을 활용하여, split으로 문자열을 : 기준으로 나눈다. 그 후, 좌우공백을 strip으로 제거하며, 비어있는 list 객체 l에 append를 사용하여, 추가한다.

2) 위와 동일한 방법이지만, 리스트 내포를 활용하여, 조건에 맞는 원소를 리스트로 초기화한다.

# 5번 문제

```python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

def addall(a): # for in 리터럴을 활용한 함수 정의
    sum = 0
    for i in range(len(list(a))):
        sum += a[i]
    return sum

print(addall(a))

def addall(a): # sum 내장함수를 활용한 리스트 내포 방식 함수 정의
    return sum(list(i for i in a))

print(addall(a))
```

1) for문을 활용한 함수정의. a를 받아서, 그 원소를 전부 더하는 단순한 방식

2) 리스트 내포방식을 활용한 함수 정의. a의 원소들을 list로 생성한 후, 이를 sum()을 통해 전부 더하여 return

```python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

def addallodd(a): # for in 리터럴을 활용한 함수 정의
    sum = 0
    for i in range(len(list(a))):
        if(a[i] % 2 == 1):
            sum += a[i]
    return sum

print(addallodd(a))

def addallodd(a): # sum 내장함수를 활용한 리스트 내포 방식 함수 정의
    return sum(list(i for i in a if(i % 2 == 1)))

print(addallodd(a))
```

1) for문을 활용한 함수정의. a를 받아서, 만약 2로 나눈나머지가 1일경우, 즉 홀수일 경우 그 원소를 전부 더하는 단순한 방식

# 7번 문제

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
items = list(d.items())

def cmp_by_key(a):
    return a[0]

def cmp_by_items(a):
    return a[1]

items.sort(key = cmp_by_key)
print(items)

items.sort(key = cmp_by_key, reverse = True)
print(items)

items.sort(key = cmp_by_items)
print(items)

items.sort(key = cmp_by_items, reverse = True)
print(items)
```

사전 d의 key와 item을 받아 list자료형으로 items리스트에 저장한다. 그 후, 시퀀스 자료형을 받아서, 그 첫번째 원소 즉 key값을 return하는 함수 cmp_by_key 함수와 item값을 return하는 함수 cmp_by_items를 정의한다. 다음에 items리스트를 sort하는데, key값으로 각 함수들을 호출하여 sort해준다. 내림차순의 경우, reverse인자를 True를 주어 오름차순을 reverse한다.

# 9번 문제

```python
from urllib import request 
 
url = 'http://cse.koreatech.ac.kr'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')

tmp = []

while(source.count('<!')):
    ann1 = source.find('<!')
    ann2 = source.find('-->')
    source = source.replace(source[ann1:ann2 + 3], '')
    
for i in range(len(source)):
    if(source[i] == '<'):
        cnt1 = i
        continue
    if(source[i] == '>'):
        cnt2 = i
        tmp.append(source[cnt1:cnt2 + 1])

for i in range(len(tmp)):
    source = source.replace(tmp[i], '')
    
print(source)
```
